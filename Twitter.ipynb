```python

from pyspark.sql import SparkSession
spark = (SparkSession
         .builder
         .appName("Python Spark SQL basic example")
         .config("spark.some.config.option", "some-value")
         .getOrCreate()
         )

```

```python
filename = '*'
df = spark.read.format('json').option('head',True).load('hdfs:///user/jlgarcia/flume/tweets/' + filename)
#df.registerTempTable("user_tt")

#info = sqlContext.sql("SELECT name FROM user_tt")
#info.show()
```

```python
#df.createOrReplaceTempView('tweets')
df.registerTempTable('tweets')
dftweetCompleto = spark.sql("SELECT * FROM tweets")
dftweetCompleto.printSchema()
```

```python
Tmp = spark.sql("SELECT lang, count(lang) as contador FROM tweets where lang <> 'und' GROUP BY lang ORDER BY contador DESC")
Tmp.show(100)

```

```python
Tmp2 = spark.sql("SELECT user.location, count(*) as contador FROM tweets GROUP BY user.location")
Tmp2.show(100)
```

```python
Tmp3 = spark.sql("SELECT place.country, place.country_code, count(1) FROM tweets group by place.country, place.country_code")
Tmp3.show(10000)
```

```python
Tmp4 = spark.sql("SELECT count(*) as contador FROM tweets")
Tmp4.show(10)
```
